Question 1

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("co2_emissions.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(10,6))
plt.plot(df["Year"], df["CO2_Emissions"], marker='o', linestyle='-', color='b', label="CO Emissions")

plt.xlabel("Year")
plt.ylabel("CO Emissions")
plt.title("Global CO Emissions Over Time")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

Question 2

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_scores.csv")
df.columns = df.columns.str.strip()

plt.figure(figsize=(8,5))
plt.plot(df["Exam_Number"], df["Score"], marker='o', linestyle='-', color='g', label="Score Progression")

plt.xlabel("Exam Number")
plt.ylabel("Score")
plt.title("Student Performance Over Exams")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
