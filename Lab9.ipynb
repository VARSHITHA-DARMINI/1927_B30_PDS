```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/content/winequality-red.csv', sep=';')
df.head()
features = ['alcohol', 'volatile acidity', 'sulphates', 'citric acid', 'density']
sns.pairplot(df[features + ['quality']], hue='quality', palette='viridis')
plt.suptitle('Pair Plot of Selected Wine Features by Quality', y=1.02)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'study_time': [5, 10, 3, 8, 6, 9, 2, 7, 4, 10],
    'attendance': [90, 95, 75, 85, 80, 92, 60, 88, 70, 97],
    'exam_score': [78, 88, 65, 83, 72, 91, 55, 85, 60, 95],
    'gender': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male']
}

df = pd.DataFrame(data)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[['study_time', 'attendance', 'exam_score']] = scaler.fit_transform(
    df[['study_time', 'attendance', 'exam_score']]
)
sns.pairplot(df, hue='gender', palette='Set2')
plt.suptitle('Student Performance Visualization', y=1.02)
plt.show()
```
