```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate housing dataset
np.random.seed(0)
data = pd.DataFrame({
    'CRIM': np.random.uniform(0.1, 10, 50),       # Crime rate
    'RM': np.random.uniform(3, 9, 50),            # Average rooms
    'AGE': np.random.randint(10, 100, 50),        # Age of building
    'DIS': np.random.uniform(1, 12, 50),          # Distance to employment centers
    'MEDV': np.random.uniform(15, 50, 50)         # Median value of homes
})

# Compute correlation matrix
corr_matrix = data.corr()

# Plot correlation heatmap
plt.figure(figsize=(8,6))
sns.set_theme(style="whitegrid")
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Housing Features", fontsize=14)
plt.show()
```

```python
# Select exam-related columns
exam_data = data[['Student','Class','Subject','Exam_Result']]

# Set theme and palette
sns.set_theme(style="darkgrid")
palette = sns.color_palette("Set2")

# Catplot (boxplot)
sns.catplot(
    data=exam_data,
    x="Class",
    y="Exam_Result",
    hue="Subject",
    kind="box",
    palette=palette,
    height=5,
    aspect=1.2
)

# Customize plot
plt.title("Exam Results by Class and Subject", fontsize=14)
plt.ylabel("Marks")
plt.xlabel("Class")
plt.grid(True)
plt.show()
```

```python
# Create example exam data
exam_data = pd.DataFrame({
    'Student': range(1, 51),
    'Class': np.random.choice(['A', 'B', 'C'], 50),
    'Subject': np.random.choice(['Math', 'Science', 'History'], 50),
    'Exam_Result': np.random.randint(40, 100, 50)
})

# Set theme and palette
sns.set_theme(style="darkgrid")
palette = sns.color_palette("Set2")

# Catplot (boxplot)
sns.catplot(
    data=exam_data,
    x="Class",
    y="Exam_Result",
    hue="Subject",
    kind="box",
    palette=palette,
    height=5,
    aspect=1.2
)

# Customize plot
plt.title("Exam Results by Class and Subject", fontsize=14)
plt.ylabel("Marks")
plt.xlabel("Class")
plt.grid(True)
plt.show()
```
