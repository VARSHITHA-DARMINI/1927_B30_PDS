Question 1

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("traffic_source.csv")

plt.figure(figsize=(8,6))
bars = plt.bar(df['Source'], df['Visits'], color='skyblue')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 20,
             str(bar.get_height()),
             ha='center', va='bottom', fontsize=10)

plt.title("Website Traffic by Source")
plt.xlabel("Traffic Source")
plt.ylabel("Number of Visits")
plt.show()

max_source = df.loc[df['Visits'].idxmax()]
print(f"Highest traffic source: {max_source['Source']} with {max_source['Visits']} visits")
```

Question 2

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("monthly_rainfall.csv")

seasons = {
    'Winter': ['December', 'January', 'February'],
    'Summer': ['March', 'April', 'May'],
    'Monsoon': ['June', 'July', 'August', 'September'],
    'Autumn': ['October', 'November']
}

colors = []
for month in df['Month']:
    if month in seasons['Winter']:
        colors.append('lightblue')
    elif month in seasons['Summer']:
        colors.append('orange')
    elif month in seasons['Monsoon']:
        colors.append('green')
    else:
        colors.append('brown')

plt.figure(figsize=(10,6))
bars = plt.bar(df['Month'], df['Rainfall'], color=colors)
plt.xticks(rotation=45)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 5,
             str(bar.get_height()),
             ha='center', va='bottom', fontsize=9)

plt.title("Average Monthly Rainfall")
plt.xlabel("Month")
plt.ylabel("Rainfall (mm)")
plt.show()

wettest = df.loc[df['Rainfall'].idxmax()]
driest = df.loc[df['Rainfall'].idxmin()]
print(f"Wettest month: {wettest['Month']} with {wettest['Rainfall']} mm rainfall")
print(f"Driest month: {driest['Month']} with {driest['Rainfall']} mm rainfall")
```
